```python
df['Previous day price'] = df['Price'].shift(1)
df
```

```python
df['Price Change'] = df['Previous day price'] - df['Price']
df
```

```python
df['5 days % return'] = (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```

```python
df.set_index('Date', inplace = True)
df = df['Price']
df
```

```python
df.index
```

```python
df.index = pd.date_range(start = '15-Aug-17', periods = 10, freq ='B')
df.index
```
