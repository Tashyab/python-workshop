```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_data.csv")
df = df.fillna({"event" : "No event"})
```

```python
df
```

```python
new_df = df.replace({"temperature":['-999', '-1'], "windspeed":['-999', '-1'], "event": "0"}, np.NaN)
new_df
```

```python
new_df = df.replace({'-999':np.NaN, '-1':np.NaN, 'No event':'Clear'})
new_df
```

```python
new_df = df.replace({'temperature': "[A-Za-z]",
                    'windspeed': "[A-Za-z]"}, '', regex=True)
new_df = new_df.replace({'-999' : np.NaN, '-1': np.NaN, 'No event': "Clear"})
new_df
```
