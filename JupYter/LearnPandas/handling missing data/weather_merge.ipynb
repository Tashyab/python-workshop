```python
import pandas as pd
temperature_df = pd.DataFrame({
    "cities": ["Patna", "Kolkata", "Guwahati", "Munich"],
    "temperature": [34, 30, 28, 16]
})
windspeed_df = pd.DataFrame({
    "cities": ["Guwahati", "Kolkata", "Patna", "Neemuch"],
    "windspeed": [8, 6, 4, 10]
})
```

```python
df = pd.merge(temperature_df, windspeed_df, on="cities", how="outer", indicator=True)
df
```

```python
new_df = df.fillna({"temperature":28, "windspeed": 4})
new_df
```

```python
another_df = pd.merge(temperature_df, windspeed_df, on='cities', how = 'outer')
```

```python
same_df = pd.DataFrame({
    "cities" : ["Patna", "Jaipur", "Kolkata"],
    "temperature" : [37, 39, 32],
    "windspeed" : [5, 8, 7]
})
df = pd.merge(same_df, df, on = 'cities')
df
```

```python
df = pd.merge(same_df, another_df, on = "cities", suffixes = ['_left', '_right'], how = 'outer')
df
```

```python
new_df = df.fillna({"temperature_left" : 31, "temperature_right" : 29, 
                    "windspeed_left":4, "windspeed_right":6})
new_df
```
